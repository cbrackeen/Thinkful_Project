```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

student_data = pd.read_csv("student-mat.csv")
```

```python
student_data.head().T
```

```python
student_age = student_data.age
```

```python
sns.set(style="ticks")
```

```python
sns.boxplot(y="age", data=student_data);
```

We can easily see the mean and one outlier. We cannot however see the distribution and see how many values are above the 75% quartile to the max.

```python
sns.swarmplot(x=student_age)
```

We can easily see the distribution and the exact spread count of how many values fall into each category, but we cannot readily see any outliers or the central tendenacy of the graph.

```python
sns.violinplot(y=student_age);
```

We can see a more contious distribution as well as the mean, inner-quartile range and the max/min values. This graph skews the data a bit, so it looks like the lowest age is 14 and highest age is 23 and does not easily identify any outliers.

```python
sns.countplot(y=student_age, palette='muted');
```

We can see the distribution, the counts that fall within each bucket, and easily tell that this data set is not normally distributed. We cannot easily tell the central tendenacy, the variance of the data, or any easily identifiable outliers (set but the IQR).

```python
sns.lmplot(x='absences', y='G3', data=student_data, x_jitter=0.6);
```

We can see the distribtion of final grades as well as the distribution of absences, as well as the correlation line between the two. It is not easy to tell the central tendenacy of either of the datasets and outliers in absences seems to skew the data so no real relationship can be seen.

```python
sns.jointplot("absences", "G3", data=student_data, kind='kde');
```

We can more easily see the continuous distributions, including central tendenacy and variance of both datasets individually and the spread of the data. We can see that there is an outlying trend at 0 for both absences and final grade, completely outside of our relationship on centering about 10-11/20 for the final grade. This graph is still highly susceptible to outliers and the data is dramatically skewed so it is hard to identify a correlation, more of a central tendenacy for correlation as well.

```python
sns.factorplot(x="absences", y="G3", data=student_data, ci=90);
```

There is too much data here, too many options that the graph is very noisy and hard to look at. Each group of data has data bars, but that is not super useful in this case since there aren't a ton of data points in the same boxes (this is not great for continuous data - almost like a bar graph). We cannot easily tell central tendenecy of absences, but can see that grades hover around the 10-12 range. This graph does not give us very useful data at all.

```python
sns.boxplot(x="Fjob", y="G3", data=student_data);
```

We can easily compare the central tendenecy of the data different groups, seeing that one Father's job seems to have more of a correlation with higher final grades. This also easily shows outliers. We cannot see how the data is distributed over all or for each group individually.

```python
sns.factorplot(x="Fjob", y="G3", data=student_data);
```

With less individual groups this grap is a lot easier to see. Not only can we see the correlation of the data, but we can see the variance in the data so we know how confidently we can compare them, like in the situation of at_home job, not very much. We cannot easily see the distributions of final grades over all or the average final grade for comparisons to the averages per group that we can see.

```python
sns.lmplot(x="age", y="G3", data=student_data, col="Fjob");
```

This graph we can easily see the relationship between age and final grade within each Fjob category. Additionally we can see the distributions as well as the actual counts in each group. With such little data, splitting it up does not show the most reliable correlations.

```python
sns.lvplot(x="Fjob", y="G3", data=student_data, palette="mako");
```

This graph we can see the comparison between the central tendency as well as the distribution of each data set and the outliers. We cannot see how many data points are in each section but can roughly see by the variance.

```python
sns.factorplot(x="Fjob", y="G3", data=student_data, kind="bar");
```

We can easily tell the difference between the different jobs and see the comparison via the error bars however this takes up a lot of space and we cannot see the distribution of the data or any outliers that may be included.

```python
sns.swarmplot(x="Fjob", y="G3", data=student_data, size=2)
sns.boxplot(x="Fjob", y="G3", data=student_data)
```

This graph we can see central tendency, variance, actual count, and outliers. We cannot see any central tendency or the distribution of all the final grades combined.
