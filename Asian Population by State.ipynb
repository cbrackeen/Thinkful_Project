```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_json("Asian_Pop_By_State.json")
```

```python
df.head()
```

```python
df.loc[1:,[1]] = df.loc[1:,[1]].astype(int)
```

```python
df.head()
```

```python
df1=df.loc[1:,[0,1]]
state_abbrev = np.array([['AL'],['AK'], ['AZ'], ['AR'], ['CA'], ['CO'], ['CT'], ['DE'], ['DC'], ['FL'], ['GA'], ['HI'], ['ID'], ['IL'], ['IN'], ['IA'], ['KS'], ['KY'], ['LA'], ['ME'], ['MD'], ['MA'], ['MI'], ['MN'], ['MS'], ['MO'], ['MT'], ['NE'], ['NV'], ['NH'], ['NJ'], ['NM'], ['NY'],['NC'], ['ND'], ['OH'], ['OK'], ['OR'], ['PA'], ['RI'], ['SC'], ['SD'], ['TN'], ['TX'], ['UT'], ['VT'], ['VA'], ['WA'], ['WV'], ['WI'], ['WY'], ['PR']])
state_abbrev2 = np.array(['','AL','AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'DC', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA','MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY','NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY', 'PR'])
```

```python
df1.columns = ["State","Population"]
df1.iloc[:,[0]] = state_abbrev
```

```python
plt.figure(figsize=(15,10))
plt.plot(df1["Population"])
plt.xlim([1,52])
plt.xticks(range(53), state_abbrev2);
```

Here we have the population of people who identify as Asian in the United States. From this graph we can see that the highest Asian population is in California, which is significantly more than any state. This fits our hypothesis as California is the closest, most accessible state in the United States relative to China.

```python
plt.figure(figsize=(10,5));

plt.subplot(1,2,1)
plt.plot(df1["State"],df1["Population"])
plt.title("Asian Population by State")
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.xlabel("States")
plt.ylabel("Population")
plt.xlim(['AL','PR'])
plt.savefig('Apop.pdf')

plt.subplot(1,2,2)
plt.scatter(df1["State"], df1["Population"])
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```

Here we can see the same data compare shown as a scatter plot. Here it seems a little more obvious that with the exception of 3 outliers, the relationship is relatively linear and constant.
