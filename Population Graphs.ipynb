```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3 = pd.read_json("Men_By_State.json")
```

```python
df4 = pd.read_json("Women_By_State.json")
```

```python
df3.head()
```

```python
df4.head()
```

```python
df5 = df3.merge(df4, on=0)
```

```python
df5.head()
```

```python
df6 = df5.iloc[1:,[0,1,3]]
```

```python
df6.columns = ["States", "Men", "Women"]
```

```python
df6.head()
```

```python
df6.loc[:,["Men","Women"]] = df6.loc[:,["Men","Women"]].astype(int)
```

```python
df6.head()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df6["Men"], color="blue")
plt.plot(df6["Women"], color="red")
plt.xticks(range(52), df6["States"], rotation='vertical');

```

The above data shows the differences in population between men and women. As we would expect the population difference isn't so drastically different state to state that it is barely noticeable on the graph.

```python
plt.figure(figsize=(15, 5))
plt.plot(((df6["Men"]-df6["Women"])/(df6["Men"]+df6["Women"])), color="blue")
plt.xticks(range(52), df6["States"], rotation='vertical');
plt.xlabel("States in America")
plt.ylabel("Percent Difference between Men and Women")

plt.title("")
#plt.savefig('testplot2.png')
```

The same data is show in this graph, but showing the differences as a function of the States. In this visualization we can more easily see the percent difference by state. Only 9 states have a positive value, indicating that overall there are a higher percentage of women in the United States.
