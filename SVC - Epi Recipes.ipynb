```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV
%matplotlib inline
```

```python
data = pd.read_csv('epi_r.csv')
```

```python
#Checking out the list of columns
list(data.columns)
```

```python
#Converting rating to binary. 3+ is good <3 is bad
data.rating = (data.rating >= 3).astype(int)
```

```python
X = data.drop(columns=['rating','title', 'calories', 'protein', 'fat', 'sodium'])
corr_list = []

#Getting correlation coefficents for each of the features with ratings
for col in X.columns:
    #Only getting a single coefficent and the name of the column
    cc = np.corrcoef(X[col],data.rating)[0][1]
    #Adding it into a sortable list
    corr_list.append([np.abs(cc),col])
```

```python
#Sort the list by the highest correlation
corr_list = sorted(corr_list, reverse=True)
```

```python
#Get list of feature names
features = [x[1] for x in corr_list[0:30]]
```

```python
#Run SVC using new list of 30 elements
svc = SVC()
X_short = data[features]
Y = data.rating

svc.fit(X_short,Y)
print(svc.score(X_short,Y))
print(cross_val_score(svc, X_short, Y))
```
