```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("WELLCOME/WELLCOME_APCspend2013_forThinkful.csv", encoding="Latin_1")
```

```python
#Taking a look at the data
data.head()
```

```python
#Capitalization not allowing correct separation. Making it lowercase and stripping
data["cleaned_title"] = (data["Journal title"].apply(lambda x: str(x).lower().strip()))
```

```python
#Value counts per top 5 journals
data["cleaned_title"].value_counts().head()
```

```python
#
data["cleaned_cost"] = (data["COST (£) charged to Wellcome (inc VAT when charged)"]
                        .apply(lambda x: float(str(x).replace('$','')
                        .replace('£', ''))))
```
