```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1 Using expoential distribution
exponetialvar = np.random.exponential([5.0], 100)
```

```python
plt.hist(exponetialvar);
plt.show()
```

```python
ex_mean = np.mean(exponetialvar)
```

```python
ex_stdev = np.std(exponetialvar)
```

```python
plt.hist(exponetialvar);

plt.axvline(ex_mean, color="r", linestyle="solid", linewidth=2);
plt.axvline(ex_mean+ex_stdev, color="r", linestyle="dashed", linewidth=2);
plt.axvline(ex_mean-ex_stdev, color="r", linestyle="dashed", linewidth=2);
plt.show()
```

Because the data is heavily skewed, the mean and the standard deviation does not describe the plot well.

```python
#2 Using power distribution
powervar = np.random.power([12], 100)
plt.hist(powervar);
plt.axvline(powervar.mean(), color="r", linestyle="solid", linewidth=2);
plt.axvline(powervar.mean()+powervar.std(), color="r", linestyle="dashed", linewidth=2);
plt.axvline(powervar.mean()-powervar.std(), color="r", linestyle="dashed", linewidth=2);
plt.show()
```

Data is still heavily skewed, so mean and standard deviation tell us a little more information about the graph than expoential, but still not the best way to describe the graph, as it is not centered.

```python
#3 Using a Trianglular distribution
trianglevar = np.random.triangular(1,10,20, 100)
plt.hist(trianglevar);
plt.axvline(trianglevar.mean(), color="r", linestyle="solid", linewidth=2);
plt.axvline(trianglevar.mean()+trianglevar.std(), color="r", linestyle="dashed", linewidth=2);
plt.axvline(trianglevar.mean()-trianglevar.std(), color="r", linestyle="dashed", linewidth=2);
plt.show()
```

Data is centrally distributed about the mean (which was one of the inputs of this distribution) and therefore the mean and standard deviation give useful descriptions of the data.

```python
#4 Using Negative Binomial Distribution
negbinomialvar = np.random.negative_binomial(100,0.7,100)
plt.hist(negbinomialvar);
plt.axvline(negbinomialvar.mean(), color="r", linestyle="solid", linewidth=2);
plt.axvline(negbinomialvar.mean()+negbinomialvar.std(), color="r", linestyle="dashed", linewidth =2)
plt.axvline(negbinomialvar.mean()-negbinomialvar.std(), color="r", linestyle="dashed", linewidth =2)
plt.show()
```

This distribution is also centrally centered about the mean, given that it is a probability graph given 100 repeititons and approaches normality. The mean and standard deviation appropriately describe the distribution

```python
#5 Using Chi Squared Distribution
chisquarevar = np.random.chisquare([3], 100)
plt.hist(chisquarevar);
plt.axvline(chisquarevar.mean(), color="r", linestyle="solid", linewidth=2)
plt.axvline(chisquarevar.mean()+chisquarevar.std(), color="r", linestyle="dashed", linewidth=2)
plt.axvline(chisquarevar.mean()-chisquarevar.std(), color="r", linestyle="dashed", linewidth=2)
plt.show()
```

The mean and standard deviation are not a great representation, as this data is skewed as well. The mean is a bit closer to the center of the data where a majority of the points lie, but is still off to the right.

```python
#6 Using geometric distribution
geometricvar = np.random.geometric(0.4, 100)
plt.hist(geometricvar);
plt.axvline(geometricvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(geometricvar.mean()+geometricvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(geometricvar.mean()-geometricvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

Mean and standard deviation do not describe this well as most of the data congregates at the 1-2 range and tapers off in one direction from there. It is heavily skewed to the right. The best graphs described by mean and standard deviation are the ones that are closest to normal and have many repetitions in their formulas.

```python
#Additional Graph
normal_graph1 = np.random.normal(5,0.5,100)
normal_graph2 = np.random.normal(10,1,100)
normalvar = normal_graph1+normal_graph2
plt.hist(normalvar);
plt.axvline(normalvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normalvar.mean()+normalvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normalvar.mean()-normalvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

Graph is approximately normal, with the mean showing an accurate central tendency and data tapering off equally in both directions. Very slightly skewed to the left.
